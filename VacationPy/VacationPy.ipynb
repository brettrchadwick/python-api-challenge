```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import gmaps
import requests
import json
from config import g_key
inputfile='../WeatherPy/output_data/cities.csv'
```

```python
#read in csv file from WeatherPy
humidity_df=pd.read_csv(inputfile)
humidity_df.head()
```

```python
#configure gmaps
gmaps.configure(api_key=g_key)
```

```python
#store lat and long in locations variable
locations=humidity_df[['Latitude','Longitude']]
#store humidity for heatmap function
humidity=humidity_df['Humidity'].astype(float)
```

```python
#set figure up and put humidity as weight for heatmaps
fig=gmaps.figure()
heat_layer=gmaps.heatmap_layer(locations, weights=humidity, max_intensity=100)
fig.add_layer(heat_layer)
```

```python
#display figure
fig
```

```python
#narrow done my ideal vacation spots by filtering out cities
hotel_df=humidity_df.loc[(humidity_df['Humidity']<40) & (humidity_df['Max Temp'] >70) &
                        (humidity_df['Max Temp']<78) &(humidity_df['Cloudiness']<50) &
                        (humidity_df['Wind Speed']<10)]

hotel_df['Hotel Name']=''
hotel_df
```

```python
# params dictionary to update each iteration
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key
}

# Use the lat/lng we recovered to identify hotels
for index, row in hotel_df.iterrows():
    # get lat, lng from df
    lat = row["Latitude"]
    lng = row["Longitude"]

    # change location each iteration while leaving original params in place
    params["location"] = f"{lat},{lng}"

    # Use the search term: "hotel" and our lat/lng
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # make request 
    response = requests.get(base_url, params=params).json()
    results=response['results']
    try:
        hotel_df.loc[index, 'Hotel Name']=results[0]['name']
    except:
        print(f'This hotel cannot be located')
```

```python
#drop any cities where a hotel could not be located
hotel_df.dropna(subset=['Hotel Name'])
hotel_df['Hotel Name'].replace('', np.nan, inplace=True)
hotel_df.dropna(subset=['Hotel Name'], inplace=True)
hotel_df
```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name, add markers to original figure
#and add layer to original fig and display fig
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations1 = hotel_df[["Latitude", "Longitude"]]
fig=gmaps.figure()
markers=gmaps.marker_layer(locations1, info_box_content=hotel_info, hover_text=hotel_info)
fig.add_layer(markers)
fig.add_layer(heat_layer)
fig
```
