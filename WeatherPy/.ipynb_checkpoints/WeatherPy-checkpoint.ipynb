For the city latitude versus max temperature, there was a correlation when linear regression was ran in the northern hemisphere. The r-squared value was -0.8 which shows a strong correlation between the data. This means as the latitude increases, the max temperature decreases. This makes sense because the closer to the equator a person is, the temperature is usually warmer than someone on the north pole. However, there was no correlation when linear regression was ran for the southern hemisphere for max temperature and latitude because the r-squared value was 0.24. This can be explained by visually looking at the points on the plot. There are some data points that skew the linear regression data. These outliers make it so a correlation cannot be made mathematically. In the dataset, there were 384 cities in the northern hemisphere, while there were only 191 in the southern hemisphere. Since the latitudes and longitudes were selected at random, this discrepancy in the number of datapoints in each set could be attributed to not being able to draw correlations. The southern hemisphere data would be more heavily weighted, so skewing of results would be more abundant with outlier data.

There was no correlation between wind speed and latitude in either the northern or southern hemispheres. The r-squared value was 0.01 and 0.2 respectively. This means that you cannot accurately predict what the wind speed for a location just given the latitude. This is true because so many factors play a role in wind speed. Differing land formations (mountains, rivers, deserts, canyons, etc.) have different wind properties and these land formations are not specific to a certain latitude.

There was not a mathematical correlation between humidity and latitude in either the northern or southern hemispheres. The r-squared value was 0.06, 0.21 respectively. This means you cannot accurately predict what the humidity for a city would be just given the latitude. This data was collected on February 26th, 2020. If it had been taken in the middle of July, the results may have been different. The humidity varies more in the summer months due to typically higher temperatures and a few other factors. There is more variation of rainfall and temperatures which have effects on humidity in the summer months. However, it was taken in February so conclusive correlations cannot be drawn. 

The randomness of the latitudes and longitudes generated could also have played a role in the results obtained from the linear regression model. Since it did not select equal numbers in both the northern and southern hemispheres, the results may be a little skewed. Also because there is a lot of open ocean in the southern hemisphere, there were not equal cities selected towards the upper and lower limits of latitude. If this experiment was performed again, I would randomly generate latitudes and longitudes with differing parameters so that more representation could be made across all continents and environments.

```python
#create northern and southern hemisphere dataframes
north=info_df.loc[info_df['Latitude'] >=0]
south=info_df.loc[info_df['Latitude'] <0]
```

```python
#run lin regress on lat vs. temp in northern hemisphere
northtemp=north['Max Temp']
northlat=north['Latitude']
plt.scatter(northlat,northtemp, marker='+')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('City Latitude vs. Max Temperature in Northern Hemisphere (2/26/2020)')
plt.xlim(north['Latitude'].min()-5, north['Latitude'].max()+5)
plt.ylim(north['Max Temp'].min()-5, north['Max Temp'].max()+5)

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(northlat, northtemp)
regress_values = northlat * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(northlat,regress_values,"r-")
plt.annotate(line_eq,(0,-5),fontsize=15,color="red")
plt.savefig('output_data/lattempnorthregress')
print(f"The r-squared is: -{round(rvalue**2,2)}")
plt.show()
#The r-squared value is -0.81, which shows a strong negative correlation between latitude and max
#temperature. As the latitude increases, the max temp decreases. This makes sense because
#the further north you go, the colder it gets. For example, the arctic is way colder than Mexico. 
```

```python
#run lin regress on max temp and latitude for southern hemisphere
southtemp=south['Max Temp']
southlat=south['Latitude']
plt.scatter(southlat,southtemp, marker='+')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('City Latitude vs. Max Temperature in Southern Hemisphere (2/26/2020)')
plt.xlim(south['Latitude'].min()-5, south['Latitude'].max()+5)
plt.ylim(south['Max Temp'].min()-5, south['Max Temp'].max()+5)

(slope1, intercept1, rvalue1, pvalue1, stderr1) = st.linregress(southlat, southtemp)
regress_values1 = southlat * slope1 + intercept1
line_eq1 = "y = " + str(round(slope1,2)) + "x + " + str(round(intercept1,2))
plt.plot(southlat,regress_values1,"r-")
plt.annotate(line_eq1,(-50,42),fontsize=15,color="red")
print(f'The r-squared value is: {round(rvalue1**2,2)}')
plt.savefig('output_data/lattempsouthregress')
plt.show()
#The r-squared value is 0.24 which for this particular dataset means there is no correlation
#from a mathematical standpoint. This can be due to the outliers and abnormal values that can
#be seen by looking at the graph. As the latitude approaches 0, there are quite a few points
#that have low max temperatures. 
```
